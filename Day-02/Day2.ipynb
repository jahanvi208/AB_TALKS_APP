```python
text=input("enter a line: ")

words=text.split()
word_count ={}

for word in words:
    word=word.lower()
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print("Word Frequency:", word_count)
```

```python
import string

text= input("enter text: ")
text= text.lower()
text= text.translate(str.maketrans('', '', string.punctuation))
text = " ".join(text.split())

print("cleaned text:", text)
```
